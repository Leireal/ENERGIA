### Solución para Access to electricity (perc of population)

En Guinea-Bissau, en el año 2000, el acceso a la electricidad era de un 0% según el Banco Mundial. Esto indica que, en ese año, prácticamente ninguna persona en Guinea-Bissau tenía acceso a la electricidad. 

En 2000, el acceso a la electricidad en Liberia era muy limitado. La infraestructura energética del país era principalmente la red eléctrica de Monrovia, con algunas líneas de distribución a los condados cercanos.
